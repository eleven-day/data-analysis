```python
# 查询当前系统所有字体
from matplotlib.font_manager import FontManager
import subprocess

mpl_fonts = set(f.name for f in FontManager().ttflist)

print('all font list get from matplotlib.font_manager:')
for f in sorted(mpl_fonts):
    print('\t' + f)
```

```python
import matplotlib.pyplot as plt
import numpy as np

eta_min = 0
eta_max = 0.01
T_max = 100
T = np.arange(T_max)
eta = eta_min + 0.5*(eta_max-eta_min)*(1+np.cos(np.pi*T/T_max))

plt.rc("font",family="Microsoft YaHei") ###修改了这一行
plt.plot(T, eta)
plt.xlabel('训练轮次')
plt.ylabel('学习率')
plt.show()
```
